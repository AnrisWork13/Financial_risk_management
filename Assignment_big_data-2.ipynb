A fraction of total capital is invested in asset the fraction is n= 8 

```python
n = 8 
d1 = 2
d2 = 7
dummystep = 10*d1*d2
for _ in range(dummystep):
    dummy = random.uniform(0,1)
Corr = np.array([[0]*n for _ in range(n)], dtype=float)
for i in range(n):
    for j in range(n):
        Corr[i][j] = (-1)**abs(i-j)/(abs(i-j)+1)
ssigma = np.array([[0]*1 for _ in range(n)], dtype = float)
# mmu is the vector Œº in the assignment
mmu = np.array([[0]*1 for _ in range(n)], dtype = float)
ssigma[0] = 2
mmu[0] = 3
for i in range(n-1):
    ssigma[i+1] = ssigma[i] + 2*random.uniform(0,1)
    mmu[i+1] = mmu[i] + 1
ddiag = np.array([[0]*n for _ in range(n)], dtype = float)
np.fill_diagonal(ddiag, ssigma)
C2 = np.matmul(np.matmul(ddiag,Corr), ddiag)
# C is the matrix C in the assignment
C = 0.5*(C2 + C2.T)
```

C is n*n matrix... which measure the covariance among the stocks created from the above code using the correlation of i th and j th terms from the (n=8) given returns.

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from gurobipy import Model, GRB
from math import sqrt

# Values of the right hand side r
r_values = np.arange(3.00, 9.25, 0.25)

# Vectors for storing results
sigma_values = []
mu_values = []

# Optimization for each r
for r in r_values:
    m = Model('portfolio')

    # Add matrix variable for the stocks
    x = m.addMVar(n)

    # Portfolio risk (squared) using the covariance matrix
    portfolio_risk = x @ C @ x
    m.setObjective(portfolio_risk, GRB.MINIMIZE)

    # Fix budget with an equality constraint
    m.addConstr(x.sum() == 1, 'budget')

    # Constraint for expected return
    m.addConstr(mmu.T @ x == r, 'expected_return')

    # Optimize model
    m.optimize()

    # Calculate sigma (risk) and mu (return) for the optimized portfolio
    sigma_values.append(sqrt(portfolio_risk.getValue()))
    mu_values.append(r)

# Find minimum risk
min_risk_index = sigma_values.index(min(sigma_values))
min_risk_sigma = sigma_values[min_risk_index]
min_risk_mu = mu_values[min_risk_index]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(sigma_values, mu_values, marker='o', label='Efficient Frontier')
plt.scatter(min_risk_sigma, min_risk_mu, color='red', label='Minimum Risk', zorder=5)
plt.text(min_risk_sigma, min_risk_mu, f'  Min Risk\n  (œÉ: {min_risk_sigma:.2f}, Œº: {min_risk_mu:.2f})', color='red')
plt.xlabel('Portfolio Risk (œÉ)')
plt.ylabel('Expected Return (Œº)')
plt.title('Efficient Frontier with Minimum Risk Point')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Re-importing necessary libraries and redefining the modified code after a reset
from gurobipy import Model, GRB
import numpy as np
import random
from math import sqrt
import matplotlib.pyplot as plt

# Initialize parameters
n = 8
d1 = 2 # Placeholder values for the digits
d2 = 7 # Placeholder values for the digits

# Generate Correlation Matrix
Corr = np.array([[0]*n for _ in range(n)], dtype=float)
for i in range(n):
    for j in range(n):
        Corr[i][j] = (-1)**abs(i-j)/(abs(i-j)+1)

# Generate sigma and mu vectors
ssigma = np.zeros((n, 1))
mmu = np.zeros((n, 1))
ssigma[0] = 2
mmu[0] = 3
for i in range(n-1):
    ssigma[i+1] = ssigma[i] + 2*random.uniform(0, 1)
    mmu[i+1] = mmu[i] + 1

# Generate diagonal matrix from ssigma
ddiag = np.diagflat(ssigma)

# Generate covariance matrix C
C2 = ddiag @ Corr @ ddiag
C = 0.5*(C2 + C2.T)

# Values of the right hand side r
r_values = np.arange(3.00, 9.25, 0.25)

# Vectors for storing results for the modified constraint
sigma_values_mod = []
mu_values_mod = []

# Optimization for each r with the modified constraint
for r in r_values:
    m = Model('portfolio_mod')

    # Add matrix variable for the stocks
    x = m.addMVar(n)

    # Portfolio risk (squared) using the covariance matrix
    portfolio_risk = x @ C @ x
    m.setObjective(portfolio_risk, GRB.MINIMIZE)

    # Fix budget with an inequality constraint
    m.addConstr(x.sum() <= 1, 'budget_mod')

    # Constraint for expected return
    m.addConstr(mmu.T @ x == r, 'expected_return')

    # Optimize model
    m.optimize()

    # Calculate sigma (risk) and mu (return) for the optimized portfolio
    sigma_values_mod.append(sqrt(portfolio_risk.getValue()))
    mu_values_mod.append(r)

# Plotting for the modified scenario
plt.figure(figsize=(10, 6))
plt.plot(sigma_values_mod, mu_values_mod, marker='x', linestyle='--', label='Modified Efficient Frontier')
plt.xlabel('Portfolio Risk (œÉ)')
plt.ylabel('Expected Return (Œº)')
plt.title('Efficient Frontier with Modified Constraint Allowing Cash Holding')
plt.legend()
plt.grid(True)
plt.show()
```
